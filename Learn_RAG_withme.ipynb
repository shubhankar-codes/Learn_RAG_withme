```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = splitter.split_documents(docs)

print("Total Chunks:", len(chunks))
```

```python
!pip install sentence-transformers

from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

chunk_texts = [chunk.page_content for chunk in chunks]

embeddings = model.encode(chunk_texts)

print(type(embedding))
print("Dimensions:", len(embedding))
print("First 10 values:", embedding[:10])
```

*NOW we are heading towards the sementic search and Extraction*

```python
!pip install scikit-learn
from sklearn .metrics.pairwise import cosine_similarity
import numpy as np
```

what we have done here is that we have first created a query and convert this into the embedings

```python
query= "what is private defence?"
query_embedding = model.encode(query)
```

```python
query_embedding.shape
```

As we have observed till now that <br>
Chunk embedding-> 384<br>
query embedding-> 384<br>

lets  compare this using cosine similarity

```python
scores = cosine_similarity(
    [query_embedding],
    embeddings
)[0]
```

```python
print(scores[:10])
```

```python
best_index = np.argmax(scores)

print("Best Index:", best_index)

print(chunks[best_index].page_content)
```

```python
top_k=5
top_indices= np.argsort(scores) [-top_k:][::-1]
```

```python
for i, idx in enumerate(top_indices):
    print(f"\nResult {i+1}")
    print("Score:", scores[idx])
    print("Page:", chunks[idx].metadata["page"])
    print(chunks[idx].page_content[:1000])
```

WE HAVE Created a RAG System for the Indian Constitution using the pdf, now lets create a reusable test function and TEST it !


```python
def search(query, k=5):
    query_embedding = model.encode(query)

    scores = cosine_similarity(
        [query_embedding],
        embeddings
    )[0]

    top_indices = np.argsort(scores)[-k:][::-1]

    results = []

    for idx in top_indices:
        results.append({
            "score": float(scores[idx]),
            "page": chunks[idx].metadata["page"],
            "text": chunks[idx].page_content
        })

    return results
```

```python
results = search("What is murder?")

for r in results:
    print("="*80)
    print("Score:", r["score"])
    print("Page:", r["page"])
    print(r["text"][:1000])
```

```python
import google.generativeai  as genai
genai.configure(api_key='Your-Private-API-Key')
```

```python
model_gemini = genai.GenerativeModel(
    "gemini-2.5-flash"
)
```

```python
def retrieve_context(query, k=5):
    query_embedding = model.encode(query)

    scores = cosine_similarity(
        [query_embedding],
        embeddings
    )[0]

    top_indices = np.argsort(scores)[-k:][::-1]

    contexts = []

    for idx in top_indices:
        contexts.append(
            chunks[idx].page_content
        )

    return "\n\n".join(contexts)
```

```python
context = retrieve_context(
    "What is murder?"
)

print(context[:2000])
```

Create a RAG Funcion & TEST

```python
def ask_rag(query):

    context = retrieve_context(query)

    prompt = f"""
You are a legal assistant.

Answer ONLY using the provided context.

Context:
{context}

Question:
{query}

Answer:
"""

    response = model_gemini.generate_content(
        prompt
    )

    return response.text
```

```python
answer = ask_rag(
    "What is murder?"
)

print(answer)
```
